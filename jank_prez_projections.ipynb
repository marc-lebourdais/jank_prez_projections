```python
import pandas as pd
import numpy as np

pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
df = pd.read_csv("georgia")
```

```python
df['expected_votes'] = df.counted_votes / (df.reported_percent / 100)
df['uncounted_votes'] = df.expected_votes - df.counted_votes
```

```python
df['opponent_percent'] = (100 - df.margin) / 2
```

```python
df['biden_percent'] = np.where(df.candidate == 'Trump', df.opponent_percent, 100 - df.opponent_percent)
df['trump_percent'] = 100 - df.biden_percent
```

```python
df['biden_uncounted_votes'] = (df.biden_percent / 100) * df.uncounted_votes
df['trump_uncounted_votes'] = (1 - df.biden_percent / 100) * df.uncounted_votes
df['biden_votes'] = (df.biden_percent / 100) * df.counted_votes
df['trump_votes'] = (df.trump_percent / 100) * df.counted_votes
```

```python
print("          Current  Uncounted  Projected\nBiden: {:10.0f} {:10.0f} {:10.0f}\nTrump: {:10.0f} {:10.0f} {:10.0f}".format(
    df.biden_votes.sum(), df.biden_uncounted_votes.sum(), df.biden_votes.sum() + df.biden_uncounted_votes.sum(),
    df.trump_votes.sum(), df.trump_uncounted_votes.sum(), df.trump_votes.sum() + df.trump_uncounted_votes.sum()
))
```
